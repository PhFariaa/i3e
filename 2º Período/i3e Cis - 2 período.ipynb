```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import sklearn.model_selection as model_selection
import sklearn.datasets as datasets
```

```python
class KNN:
    #Definindo numeros de vizinhos e elementos de treinamento
    def __init__(self, k, X_train, y_train):
        self.k = k 
        self.X_train = X_train
        self.y_train = y_train
    #Calcular a distÃ¢ncia euclidiana
    def distancia(self, n1, n2):
        distance = np.linalg.norm(n1-n2)
        return distance
    
    #Retornar os K vizinhos mais prÃ³ximos de value
    def get_neighbors(self, value):
        distances = []
        for x in self.X_train:
            distances.append(self.distancia(value,x))
        distances = np.asarray(distances)
        indices = np.argpartition(distances, self.k)
        k_first = indices[:self.k]
        return k_first
    
    #Realizando a previsÃ£o de um elemento dado o conjunto de treinamento
    def predict(self, value):
        knn_indices = self.get_neighbors(value)
        knn_labels=[]
        for i in knn_indices:
            knn_labels.append(self.y_train[i])
        occurrences = np.bincount(knn_labels)
        node=np.argmax(occurrences)
        return node
    
    #Retornar a acurÃ¡cia
    def accuracy(self, y_test, predictions):
        correct = y_test == predictions
        acc = (np.sum(correct) / y_test.shape[0])*100
        return acc
```

```python
X,y = datasets.load_iris(return_X_y=True) #Carregando o dataset iris
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.3) #Dividindo o dataset em treinamento e teste
knn = KNN(10, X_train, y_train) #Instanciando o knn

pred=[] #VariÃ¡vel que vai armazenar o resultado da previsÃ£o

for x in X_test:
    pred.append(knn.predict(x)) # Realizando a previsÃ£o em cada elemento de X_test e guardando em preds

precisao = knn.accuracy(y_test, pred) #Comparando resultados
print(f"Precisao de {precisao}% (Minha classe)")

#Comparando com o modelo do sklearn

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knnSKL = KNeighborsClassifier(n_neighbors=10)
knnSKL.fit(X_train, y_train)
y_pred = knnSKL.predict(X_test)
precisao = metrics.accuracy_score(y_test, y_pred)*100
print(f"Acuracia de {precisao}% (Utilizando o KNN do SKLEARN)")
```

```python
#precisao quando os dados sao normalizados
for c in vinhos.columns:
    vinhos[c] = (vinhos [c] - min(vinhos[c]))/(max(vinhos[c]) -min(vinhos[c]))
    
clf = KNeighborsClassifier(n_neighbors = 10)

X = vinhos.copy().drop('wine_is_red', axis=1) #meus dados
y = vinhos['wine_is_red']#oq eu quero prever

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 13)

#Aplicando o KNN para a classificaÃ§Ã£o binaria do vinho ser tinto ou nao
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

#Vendo a precisao do modelo
precisao = metrics.accuracy_score(y_test, y_pred)*100
print(f"precisao de {precisao}%")
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
#previsao da qualidade dos vinhos com KNN
vinhos = pd.read_csv("winequality.csv")

clf = KNeighborsClassifier(n_neighbors = 5)

vinhos = vinhos.drop('Unnamed: 0', axis = 1)
X = vinhos.copy().drop('wine_is_red', axis=1) #meus dados
y = vinhos['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 13)
                          
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)*100
print(f"AcurÃ¡cia do modelo: {accuracy:.2f}")
```

```python
#Oversampling
from imblearn.over_sampling import RandomOverSampler
vinhos = pd.read_csv("winequality.csv")

vinhos = vinhos.drop('Unnamed: 0', axis = 1)
X = vinhos.copy().drop('wine_is_red', axis=1)
y = vinhos['quality']

over_sampler = RandomOverSampler()
X_resampled, y_resampled = over_sampler.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state = 13)

clf = KNeighborsClassifier(n_neighbors = 5)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)*100
print(f"AcurÃ¡cia do modelo: {accuracy:.2f}")

print("Melhora consideravel na acuracia do modelo, sem o oversampling o KNN estava com acuracia de 53-55%%")
```

```python
#Undersampling
from imblearn.under_sampling import RandomUnderSampler
vinhos = pd.read_csv("winequality.csv")

vinhos = vinhos.drop('Unnamed: 0', axis = 1)
X = vinhos.copy().drop('wine_is_red', axis=1)
y = vinhos['quality']

under_sample = RandomUnderSampler()
X_resampled, y_resampled = under_sample.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state = 13)

clf = KNeighborsClassifier(n_neighbors = 5)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)*100
print(f"AcurÃ¡cia do modelo: {accuracy:.2f}")

print("Piora drastica na acuracia do modelo")
```
